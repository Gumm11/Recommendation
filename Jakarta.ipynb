```python
from sklearn.cluster import DBSCAN
from sklearn.metrics.pairwise import haversine_distances
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors
import hdbscan
from datetime import datetime, timedelta
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import joblib
import random
```

```python
df = pd.read_csv('jakarta/TempatWisataJakarta.csv')
df.head()
```

```python
#Mengambil Variabel Yang akan Digunakan
df_jakarta= df.dropna(subset=['Reviews', 'City', 'Latitude', 'Longitude'])
df_jakarta = df_jakarta[df_jakarta['State'] == "Jakarta"]

df_jakarta = df_jakarta[['Title', 'Reviews', 'Review_points', 'Latitude', 'Longitude']]

df_jakarta.head()
```

# Train The Model

```python
coords = df_jakarta[['Latitude', 'Longitude']].values
scaler = StandardScaler()
coords_scaled = scaler.fit_transform(coords)
```

```python
coords_rad = np.radians(coords)

clusterer = hdbscan.HDBSCAN(min_cluster_size=2, min_samples=1, cluster_selection_epsilon=0.00009)
cluster_labels = clusterer.fit_predict(coords_rad)

df_jakarta['Cluster'] = cluster_labels
```

```python
n_clusters = len(set(cluster_labels)) - (1 if -1 in cluster_labels else 0)
n_noise = (cluster_labels == -1).sum()

print(f"Number of clusters: {n_clusters}")
print(f"Number of noise points: {n_noise}")

# Points per cluster
print(df['Cluster'].value_counts())
```

```python
plt.figure(figsize=(20, 8))
for cluster_id in sorted(df_jakarta['Cluster'].unique()):
    label = f"Cluster {cluster_id}" if cluster_id != -1 else "Noise"
    cluster_points = df_jakarta[df_jakarta['Cluster'] == cluster_id]
    plt.scatter(cluster_points['Longitude'], cluster_points['Latitude'], label=label, s=30)

plt.title("HDBSCAN Clustering")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.legend()
plt.show()
```

```python
clusters = {}
for cluster_id in sorted(df['Cluster'].unique()):
    if cluster_id == -1:
        continue
    clusters[f"Cluster {cluster_id + 1}"] = df[df['Cluster'] == cluster_id][['Title', 'Latitude', 'Longitude']]

for day, locations in clusters.items():
    print(f"\n{day} ({len(locations)} locations):")
    print(locations)
```

```python
joblib.dump(clusterer, 'jakarta.pkl')
print("Model saved successfully!")
```
